```python
import geopandas

cadaster = geopandas.read_file("../Data/SHP/Lots/Full_Seigneury_Merged_v2.shp")
cadaster["ORIGINAL_A"] = cadaster["ORIGINAL_A"].str.capitalize()
renamed = cadaster.rename(columns = {"YEAR_1": "year"})
new_cadaster = renamed.to_crs(epsg=4326)
new_cadaster.to_file('../Data/GEOJSON/Full_Cadaster.geojson', driver = 'GeoJSON')


augmentations = geopandas.read_file("../Data/SHP/Augmentation/Seigneurial_augmentations.shp")
new_augmentations = augmentations.to_crs(epsg=4326)
new_augmentations.to_file('../Data/GEOJSON/augmentations.geojson', driver = 'GeoJSON')

kanehsatake = geopandas.read_file("../Data/SHP/Kanehsatake/Kanehsatake.shp")
new_kanehsatake = kanehsatake.to_crs(epsg=4326)
new_kanehsatake.to_file('../Data/GEOJSON/kanehsatake.geojson', driver = 'GeoJSON')
```
